```python
import numpy as np
np.set_printoptions(threshold=np.nan)

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
X1 = np.random.choice(np.arange(1, 7), 10000)
X2 = np.random.choice(np.arange(1, 7), 10000, p=[1/16, 1/16, 3/16, 3/16, 4/16, 4/16])
```

```python
Y = X1*X2
e_Y = np.mean(Y)
var_Y = np.var(Y)

fig = plt.figure()
ax = fig.add_subplot(111)
p1 = sns.countplot(Y)
plt.title('Distribution of Y', size=18, y=1.03)
plt.xlabel('Y', size=16)
plt.ylabel('Count', size=16);

print('e_Y: ', e_Y)
print('var_Y:', var_Y)
```
